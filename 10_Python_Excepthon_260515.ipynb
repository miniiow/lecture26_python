```python
# try...except...
try:
    number = int(input('>> 숫자입력 : '))
    print(number)
except ValueError:
    print('숫자를 입력해주세요.')
```

```python
# except ... as ...
try:
    num_list = [2,5,34,65,11]
    index = int(input('인덱스 입력 : '))
    print(num_list[index])
except (ValueError, IndexError):
    print(f'0부터 {len(num_list)-1}사이의 정수를 입력하세요.')
# except IndexError:
#     print(f'0부터 {len(num_list)-1}사이의 정수를 입력하세요.')
except Exception as e:
    print(f'예외가 발생했습니다. : {e} -> {type(e)}')
```

```python
try:
    number= int(input('숫자 입력 : '))
    # print(number)
except ValueError:
    print('숫자를 입력하지 않았습니다.')
else:
    # 예외가 발생하지 않았을때 실행되는 코드
    print(number)
finally:
    print('끝났당 !!')
```

```python
while True:
    try:
        number = int(input('양의 정수 입력 : '))
        if number <= 0:
            # ValueError발생시 아래 except안에 있는 코드 실행한다
            raise ValueError
    except ValueError:
        print('양의 정수를 입력하세요.')
    else:
        sum = 0
        for i in range(1, number+1):
            sum += i
        print(f'>> 1부터 {number}까지의 합 : {sum}')
        break
```

## 학습내용 되돌아보기 (1)

```python
# 1)
try:
    # ValueError : 잘못된 값 입력
    f = float('30f')
except Exception as e:
    print(type(e))
```

```python
# 2)
try:
    # KeyError : 딕셔너리에 존재하지 않는 Key로 접근함
    {'fb':11,'bb':9,'vb':6}['foot']
except Exception as e:
    print(type(e))
```

```python
# 3)
try:
    # IndexError : [1,2,3] List의 3번째 인덱스 값을 가져와라
    [1,2,3][3]
except Exception as e:
    print(type(e))
```

```python
# 4)
try:
    # TypeError : 다른 타입을 변수 안에 넣을 수 없음
    pl = 'python' + 3
except Exception as e:
    print(type(e))
```

## 학습내용 되돌아보기 (2)

```python
try:
    m = int(input('월(month) 입력 > '))
    if not (1 <= m <= 12):
        raise Exception('1 ~ 12 사이의 정수를 입력하세요.')
except Exception as e:
    print(f'예외발생 : {e}')
else:
    print(f'{m}월: 잘 입력했습니다.')
```

## 학습내용 되돌아보기 (3)

```python
def divide(x, y):
    return x/y

try:
    x = float(input('x 입력 : '))
    y = int(input('y 입력 : '))
    divide(x, y)
except:
    print('0으로는 나눌 수 없습니다.')
else:
    print(f'결과 : {divide(x, y)}')
```

## 학습내용 되돌아보기 (4)

```python
uyear = {1:'freahman', 2:'sopjomore', 3:'junior', 4:'senior'}

try:
    user_input = int(input('대학교 몇 학년이지요? '))
    if not (1 <= user_input <= 4):
        raise Exception
except Exception as e:
    print(f'예외 발생 이름: {type(e)}')
    print(f'예외 발생 이유: 1~4 정수를 입력하세요.')
else:  # 예외가 발생하지 5않았을때만 실행
    print(f'{user_input}학년: {uyear[user_input]}')
finally:  # 예외 발생여부와 상관없이 무조건 실행
    print('예외 처리가 잘되는군요!')
```
