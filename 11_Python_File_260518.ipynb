```python
filename = './sample.txt'

# with구문을 사용하면 자동으로 파일을 closed한다
with open(filename, encoding='utf-8') as f_read:
    content = f_read.read()
print(content)
# ValueError: I/O operation on closed file : with구문때문에 자동으로 파일 닫혀서 읽을 수 없음
# content2 = f_read.read()
```

```python
with open(filename, encoding='utf-8') as f_read:
    lines = f_read.readlines()

for line in lines:
    print(line.strip())
```

```python
with open(filename) as f_read:
    while True:
        line = f_read.readline()
        if line: print(line.strip())
        else: break
        
```

```python
import math as m
import statistics as s

if __name__ == '__main__':
    print('1! = ', m.factorial(1))
    print('6! = ', m.factorial(6))
    print('11! = ', m.factorial(11))
    print('16! = ', m.factorial(16))

    st = [80,99,77,65,92,74,82]

    print()
    print(st)
    print(f"중앙값: {s.median(st):.2f}")
    print(f"평균: {s.mean(st):.2f}")
    print(f"분산: {s.variance(st):.2f}")
    print(f"표준편차: {s.stdev(st):.2f}")
```

```python
f = open('pyzenex.txt', 'wt')
f.write('파이썬 철학\n')
f.write('아름다움이 추한 것보다 낫다.\n')
f.write('명시적인 것이 암묵적인 것보다 낫다.\n')
f.write('단순함이 복잡한 것보다 낫다.\n')

f = open('pyzenex.txt')
print(f.read())
f.close()
print()
f = open('pyzenex.txt')
for line in f.read():
    print(line, end='')
f.close()
print()
f = open('pyzenex.txt')
for line in f:
    print(line, end='')
f.close()
print()
f = open('pyzenex.txt')
while True:
    line = f.read()
    if not line: break
    print(line, end='')
f.close()
```

### 학습 내용 되돌아보기 (3)

```python
cp = open('cp.txt', 'wt')
cp.write('freshman\n')
cp.write('sophomore\n')
cp.write('junior\n')
cp.write('senior\n')
```

```python
cp = open('cp.txt')

# enumerate() : 번호 + 값 반환하는 파이썬 내장함수
for i, line in enumerate(cp):
    print(f'{i+1}학년 {line.strip()}')

cp.close()
```

### 이미지 파일 복사하기 함수

```python
import os

def copy_image_file(image_file):
    with open(image_file, mode='rb') as f:
        source = f.read()

        name, ext = os.path.splitext(image_file)
        new_name = name + '_copy' + ext

        with open(new_name, mode='wb') as copy:
            copy.write(source)
```
