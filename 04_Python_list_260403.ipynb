### 리스트(list)만들기
*대괄호([]) 안에 데이터를 콤마(,)로 구분하여 나열한다
    - 각 데이터를 항목(item) 또는 요소(element)라고 부른다*

*비어있는 리스트의 생성*
> #### [ ]  
    빈 대괄호로 빈 리스트 생성

> #### list()  
     인자가 없는 내장함수 list()로도 빈 리스트 생성 (함수를 불러오기 때문에 [ ]보다 연산을 더 사용한다)

```python
a = []
b = list()
c = [1,2,3]
d = ['banana','apple','tomato','watemelon']
e = [1,2,3,'one','two','three']
f = [1,2,['banana', 100], ['tomato',300]]

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)

print(f[0])
print(f[1])
print(f[2])
print(f[3])
```

### 리스트의 항목에 접근하기
**인덱싱**  
*항목의 위치 인 인덱스를 사용하여 리스트의 특정 위치에 있는 항목에 접근 하는 것*
> #### list[index]

**for문으로 접근하기**  
*리스트 명에 있는 항목을 하나씩 변수에 할당하여 문장 반복 실행*
> #### for 변수 in 리스트 : 반복문장

<br>

```python
test_list = [1,2,3,'one','two','three']
print(test_list[0])
print(test_list[3])
print(test_list[-1])
print(test_list[-3])

for item in test_list:
    print(item)
```

### 리스트의 항목에 접근하기 : 슬라이싱(slicing)
*대괄호와 콜론(:)을 이용해서 리스트의 일부 항목을 추출하는 것*

> **list[start : stop : step]**

<br>

```python
test_list = [1,2,3,4,5,6,7,8]

# 생략시 이렇게 인식한다 (start : 0부터 / stop : 끝까지 / step : 1개씩)
print(test_list[0:5])
print(test_list[:5])

print(test_list[5:])
print(test_list[1::2])

print(test_list[:5:2])

print(test_list[::])
print(test_list[:])

print(test_list[-1:-7:-2])
print(test_list[1:-1])
print(test_list[-3::-1])
print(test_list[::-1])
print(test_list[::-2])
```

```python
[1,2,3][::-1]
```

```python
test_list = [1,2,3,4,5,6,7,8]
test_list[:2] = [111,222,333]
print(test_list)

# 리스트 1, 2 자리의 값을 111, 222, 333으로 대체해라
```

```python
test_list.reverse()
print(test_list)
```

### 리스트에 항목 추가하기

> #### **list.append(item)**
*리스트의 맨 끝에 항목을 추가*  

<br>

> #### **list.insert(index,item)**
*리스트의 index위치에 item을 추가 / 원래 있던 항목은 뒤로 하나씩 밀린다*  

<br>

> #### **list.extend(list2)**
*list2에 있는 항목을 list1에 추가 / list2는 반드시 리스트이어야 함*  

<br>

> #### **list3 = list1 + list2**
*두 개의 list를 연결하여 하나의 리스트 생성*  

<br>

> #### **list * number / number * list**
*list의 항목을 number만큼 반복하여 리스트 생성*  


```python
a_list = [1,2,3,4]
print('a_list : ',a_list)
a_list.append('four')
print('a_list.append("four") : ',a_list)
a_list.insert(3, 'three')
print('a_list.insert(3, "three") : ',a_list)
a_list.extend([10,11])
print('a_list.extend([10,11]) : ',a_list)

b_list = [1,2,3]
print('b_list : ',b_list)
c_list = ['a','b','c']
print('c_list : ',c_list)
d_list = b_list + c_list
print('d_list = b_list + c_list : ',d_list)

e_list = b_list * 3
print('e_list = b_list * 3 : ',e_list)
```

### 리스트에서 항목 삭제하기

> #### **data = list.pop(index)**
*list의 index에 있는 항목을 삭제하고 반환*  

<br>

> #### **data = list.pop()**
*list의 마지막 항목을 삭제하고 반환*  

<br>

> #### **list.remove(item)**
*list에서 item을 찾아서 삭제 / list에 item이 여러 개 있는 경우 첫 item만 삭제*  

<br>

> #### **del list[index] / del list[slicing]**
*list의 index 또는 slcing 범위의 항목을 삭제*  

<br>

> #### **list.clear()**
*list의 항목을 모두 삭제하여 빈 리스트로 만듦*  

<br>


```python
test_list = [1,2,3,'a','b','c',10,11]
print("test_list : ",test_list)

item = test_list.pop(2)
print("item = test_list.pop(2) : ",test_list)
item = test_list.pop()
print("item = test_list.pop() : ",test_list)
test_list.remove('a')
print("test_list.remove('a') : ",test_list)

del test_list[2]
print("del test_list[2] : ",test_list)
del test_list[:2]
print("del test_list[:2] : ",test_list)

test_list.clear()
print("test_list.clear() : ",test_list)
```

### 리스트의 항목 재배열하기

> #### **list.reverse()**
*list의 항목을 역순으로 뒤집음*  

<br>

> #### **list.sort(reverse = False, key=None)**
> #### **list2 = sorted(list1, reverse=False, key=None)**
*list의 항목을 정렬*  
*기본적으로 오름차순으로 정렬 / 내림차순 : reverse = True*  
*key=옵션으로 정렬 기준 지정 가능*  
*list.sort()는 in-place sort : 자기정렬(새로운 리스트를 만드는 것이 아님)*  
*sorted()는 정렬된 결과를 새로운 리스트로 생성함*  

<br>


```python
a_list = ['one','two','three','four']
print("a_list : ",a_list)
a_list.reverse()
print("a_list.reverse() : ",a_list)
a_list.sort()
print("a_list.sort() : ",a_list)

a_list.sort(reverse=False,key=None)
print("a_list.sort(reverse=False,key=None) : ", a_list)
a_list.sort(reverse=True,key=None)
print("a_list.sort(reverse=True,key=None) : ", a_list)
```

### 리스트 탐색하기

> #### **item in list**
*list에 item이 있는지 확인하여 있으면 True, 없으면 False를 반환*  


> #### **list.index(item)**
*list 내 item의 index를 반환, 없으면 ValueError 발생*  

> #### **list.count(item)**
*list 내 tem의 개수를 반환*  


> #### **len(list)**
*list를 구성하는 항목 수 계산*  


> #### **max(list)**
*list항목 중 최댓값 반환*  


> #### **min(list)**
*list항목 중 최소값 반환*  


> #### **sum(list)**
*list 항목들의 합 반환*  

<br>


```python
test_list = [1,2,3,1,2,3]

target = 3
print(target in test_list)
print(test_list.index(target))
print(test_list.count(target))

target = 4
print(target in test_list)
print(test_list.count(target))
print(test_list.index(target))
```

```python
a_list = ['one','two','three','four']
print(a_list)
a_list.reverse()
print(a_list)

a_list.sort()
print(a_list)

a_list = ['one','two','three','four']
b_list = sorted(a_list)
print('원본 데이터',a_list)
print('정렬 후 데이터',b_list)

c_list = sorted(a_list, key=len)
print(c_list)
```

```python
scores = [44,55,62,56, 98]

print(sorted(scores))
print(sorted(scores, reverse=True))
```

```python
# 1 ~ n까지의 수로 이루어진 리스트 생성

N = int(input('숫자 : '))
num_list = []

for i in range(1, N+1):
    num_list.append(i)
print(num_list)
```

```python
# 리스트 내포, List comprehension
# expression 사용

N = int(input('숫자 : '))
num_list = [i*2 for i in range(1, N+1)]
print(num_list)

N = int(input('숫자 : '))
num_list = [i for i in range(1, N+1, 2)]
print(num_list)

# if문 사용
N = int(input('숫자 : '))
num_list = [i for i in range(1, N+1) if i%2 == 0]
print(num_list)

N = int(input('숫자 : '))
num_list = [i for i in range(2, N+1, 2)]
print(num_list)
```
