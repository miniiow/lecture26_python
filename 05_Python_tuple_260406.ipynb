### 튜플(tuple)의 개념

*여러개의 데이터를 순서대로 저장하지만, 변경은 할 수 없는 자료구조*  
*문자열(str), 리스트(list), 튜플(tuple) 모두 시퀀스이다*  

**특징**
> 항목간에 순서가 있음 (ordered) → 항목별 위치가 정해져 있음(indexing)  <br>
> 항목 자료 값은 중복 되도 무관  <br>
> 수정 불가능 (immutable) → 항목을 추가, 삭제, 수정하려고 하면 오류 발생  <br>
> 항목은 정수, 실수, 문자열, 리스트 등이 모두 가능  <br>

**튜플의 표현**
> #### **변수명  = (item0, item1, item2 ..., item n-1)**

```python
t1 = (1,2,3)
t2 = ('a','b', ('ab','cd'))
t3 = ('tuple', (1, 2), [3,4])

print(t1)
print(t2)
print(t3)

for i in t3:
    print(i)
```

### 튜플(tuple)만들기  

**빈 튜플 생성**
> #### **(), tuble()**  <br>

**항목이 1개인 튜플의 생성**
*항목 뒤에 쉼표(,)를 반드시 붙여야 함 → 쉼표가 없으면 일반 괄호로 인식*

**튜플 표현시 괄호 생략 가능**  <br>  
**tuple packing**
*괄호 없는 튜플의 생성 : 여러개의 값을 쉼표로 구분하여 변수에 대입하면, 파이썬이 자동으로 튜플로 묶어즘*  <br><br>
**튜플 언패킹(tuple unpacking)**  
*튜플의 값을 여러 변수에 나누어 담는 것*

```python
t4 =()
print(t4)
t5 = (1,)
print(t5)
t6 = 1,2,3
print(t6)
t7 = 1,
print(t7)

x, y, z = t6
print(x, y, z)
```

```python
# 문자열, 튜플, 리스트 모두 시퀀스 맞는지 확인해보기

s = 's1,s2,s3,s4'
t = ('t1','t2','t3','t4')
l = ['l1','l2','l3','l4']

for i in s:
    print(i)

for i in t:
    print(i)

for i in l:
    print(i)
```

### 튜플(tuple) 다루기

**튜플 항목에 접근**  
*인덱싱, 슬라이싱, for...in*  <br>

**튜플 탐색**  
*in 연산자, index(), count()*  <br>

**튜플 관련 정보**  
*len(), max(), min(), sum()*  <br>

**새로운 튜플 생성**  
*sored(), + 연산, \*연산*  
*sored()는 새로운 값을 만들기 때문에 사용가능하다*  <br>

**튜플 전체 삭제**  
*del 튜플이름*  <br>

**리스트에서는 사용하지만, 튜플에서 사용할 수 없는 함수**  
*튜플은 값을 수정할 수 없기때문에 기존의 값을 수정하는 함수들을 사용할 수 없다*
> *항목 추가 : append(), insert(), extend()*  <br>
> *항목 삭제 : pop(), remove(), clear(), del 항목*  <br>
> *항목 재배열 : reverse(), sort()*  <br>

```python
t = (10,50,23,44,68,145,1,20,5)
```

```python
# 튜플 항목에 접근
print('\n튜플 항목에 접근')
print(t[0])
print(t[1:2])
print(t[1:3])
print(t[-1:-5:-2])
target = 4
for i in range(len(t)):
    if target == i:
        print(t[i])
```

```python
# 튜플 탐색
print('\n튜플 탐색')
print(24 in t)
print(t.index(1))
print(t.count(2))
```

```python
# 튜플 관련 정보
print('\n튜플 관련 정보')
print(len(t))
print(max(t))
print(min(t))
print(sum(t))
```

```python
# 새로운 튜플 생성

print('\n새로운 튜플 생성')
t2rd = tuple(sorted(t, reverse=False, key = None))
print(t2rd)
# 같은 데이터타입끼리만 연산 가능 : 튜플 +연산 하려면 tuple()로 형변환 해줘야 한다
print(t + t2rd)
print(t * 2)
```

```python
# 튜플 전체 삭제

print('\n튜플 전체 삭제')
del t2rd
# NameError: name 't2rd' is not defined
# print(t2rd)
```

```python
# 튜플에서 사용할 수 없는 함수 검증
print('\n튜플에서 사용할 수 없는 함수 검증')

# 항목추가
print(r'''
AttributeError: 'tuple' object has no attribute 'append'
t.append(12)
''')
print(r'''
AttributeError: 'tuple' object has no attribute 'insert'
t.insert()
''')
print(r'''
AttributeError: 'tuple' object has no attribute 'extend'
t.extend(2)
''')

# 항목삭제
print(r'''
AttributeError: 'tuple' object has no attribute 'pop'
t.pop()
''')
print(r'''
AttributeError: 'tuple' object has no attribute 'remove'
t.remove()
''')

```

### 튜플의 활용

**두 변수의 값 바꾸기 (swap)**

```python
x, y = 10, 20
print(x, y)
x, y = y, x
print(x, y)
```

**함수에서 여러 개의 값 반환**  
*함수에서 여러 개의 값을 한꺼번에 반환할 때 자동으로 튜플로 묶어 전달*

```python
def add_and_mul(a,b):
    return a+b, a*b

result = add_and_mul(3, 5)
print(result)
```

**데이터 무결성**  
*프로그램이 실행되는 동안 값이 절대 변하면 안 되는 경우*  <br>  
**딕셔너리의 키**  
*리스트는 딕셔너리의 키로 쓸 수 없음(리스트는 수정이 자유롭기 때문)*  <br>  
**속도 최적화**  
*단순히 읽기만 한다면 튜플이 리스트보다 메모리와 속도 면에서 유리하다*
*리스트는 동적배열이기때문에 제한이 없어 그 뒤에 올 값을 항상 대기하고 있기 때문*  <br>
