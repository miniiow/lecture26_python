>   *자식클래스: 차단*  
>   *클래스 외부: 차단*  <br>

````python
class Circle:
    pi = 3.14
    def __init__(self, radius):     # radius: 반지름
        self.radius = radius

    def area(self):
        return self.radius ** 2 * Circle.pi

c = Circle(2.5)
print(c.area())
````

````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
    
rec = Rectangle(2.1, 3.4)
print(rec.area())
````

````python
class Account:
    def __init__(self):
        self.balance = 0
    def deposit(self, amount):  # 입금
        self.balance += amount
    def withdraw(self, amount): # 출금
        self.balance -= amount
    
a1 = Account()
a1.deposit(100)

a2 = Account()
a2.deposit(50)

print('a1 : ', a1.balance)
print('a2 : ', a2.balance)
````

````python
class Account:
    def __init__(self):
        self.__balance = 0
    def deposit(self, amount):  # 입금
        self.__balance += amount
    def withdraw(self, amount): # 출금
        self.__balance -= amount
    def get_balance(self):
        return self.__balance
    
a1 = Account()
a1.deposit(100)

a2 = Account()
a2.deposit(50)

print('a1 : ', a1.get_balance())
print('a2 : ', a2.get_balance())

# Error Chk : private 변수라서 접근할 수 없음 > get메서드로 접근할  > 캡슐화
# AttributeError: 'Account' object has no attribute '_Account__balance'
````
