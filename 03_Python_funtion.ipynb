```python
# 구구단 출력 함수

# 함수정의 후에 함수 불러와야 한다
# gugudan(5)

def gugudan(num):
    print(f'\n[구구단 {num}단]')
    for i in range(1, 10) :
        print(f'{num} x {i} = {num*i}')

gugudan(5)
```

```python
number = int(input('단 입력: '))
gugudan(number)
```
